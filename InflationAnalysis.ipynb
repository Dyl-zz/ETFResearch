```python
import pandas as pd
import matplotlib.pyplot as plt

smh_inflation = pd.read_excel('InflationSMH.xlsx')
qqq_inflation = pd.read_excel('InflationQQQ.xlsx')
spy_inflation = pd.read_excel('InflationSPY.xlsx')

smh_inflation['Date'] = pd.to_datetime(smh_inflation['Date'])
qqq_inflation['Date'] = pd.to_datetime(qqq_inflation['Date'])
spy_inflation['Date'] = pd.to_datetime(spy_inflation['Date'])

smh_inflation.sort_values('Date', inplace=True)
qqq_inflation.sort_values('Date', inplace=True)
spy_inflation.sort_values('Date', inplace=True)

# Calculate statistics
datasets = {'SMH': smh_inflation, 'QQQ': qqq_inflation, 'SPY': spy_inflation}
statistics = {}

for name, data in datasets.items():
    stats = {
        'Highest Price': data['Close'].max(),
        'Lowest Price': data['Close'].min(),
        'Mean Price': data['Close'].mean(),
        'Median Price': data['Close'].median(),
        'Standard Deviation': data['Close'].std()
    }
    statistics[name] = stats

# Display statistics
statistics_df = pd.DataFrame(statistics).T
print(statistics_df)

# Plot the data
plt.figure(figsize=(12, 6))

# SMH Line Graph
plt.plot(smh_inflation['Date'], smh_inflation['Close'], label='SMH (Semiconductors)', linewidth=2)

# QQQ Line Graph
plt.plot(qqq_inflation['Date'], qqq_inflation['Close'], label='QQQ (NASDAQ-100)', linewidth=2)

# SPY Line Graph
plt.plot(spy_inflation['Date'], spy_inflation['Close'], label='SPY (S&P 500)', linewidth=2)

# Customize the plot
plt.title('Stock Prices Over Time During Inflation Period', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
